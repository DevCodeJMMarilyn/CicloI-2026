## Ejemplo 2 de Estadística robusta

Generar una muestra aleatoria con una distibución uniforme en el intervalo $[10,50]$ de un tamaño de $n=90$. Agregar en la primera posición el valor atípico 250.


a) Calcular la media recortada de la muestra al 15\%, al 20\% y calcula la centrimedia

b) Calcular la media winsorizada al 15%.

c) Calcular la trimedia

d) Compara los valores obtenidos en los apartados a), b) y c) con la media muestral.

```python
import numpy as np
import pandas as pd

a,b,n = 10,50,90 #Parámetros de la distribución normal

np.random.seed(seed=2026) #Semilla para obtener el mismo resultado
datos = np.random.uniform(a,b,n)
```

```python
datos[0] = 250 #Agregar valor atípico
datos.sort()
```

```python
datos
```

a) Calcular la media recortada de la muestra al 15 %, al 20 % y calcula la centrimedia.

```python
#Media recortada al 15%
alfa = 0.15 #1/n no es múltiplo de 0.15

#Cantidad de elementos a eliminar
k = int(n*alfa)
k
```

```python
#Eliminar k elementos a cada extremos
datos_2 = datos[k:n-k]
#Ponderar con p = 1+[alfa*n]-alfa*n el primer y último valor
p = 1+int(alfa*n)-alfa*n

datos_2[0], datos_2[-1] = p*datos_2[0], datos_2[-1]
media_recortada_15 = sum(datos_2)/(n-2*k)

print("La media recortada al 15%% es %s" %media_recortada_15)
```

```python
#Media recortada al 20%
alfa_2 = 0.20 #1/n no es múltiplo de 0.20

#Cantidad de elementos a eliminar
k = int(n*alfa_2)
k
```

```python
#Eliminar k elementos a cada extremos
np.random.seed(seed=2026) #semilla para obtener el mismo resultado
datos= np.random.uniform(a,b,n)
datos.sort()

datos_3 = datos[k:n-k]
#Ponderar con p = 1+[alfa*n]-alfa*n el primer y último valor
p = 1+int(alfa*n)-alfa*n

datos_3[0], datos_3[-1] = p*datos_3[0], datos_3[-1]
media_recortada_20 = sum(datos_3)/(n-2*k)

print("La media recortada al 20%% es %s" %media_recortada_20)
```

```python
alfa = 0.25

k = int(alfa*n)
datos_4 = datos[k:n-1-k]
p = 1+int(alfa*n)-alfa*n
datos_4[0] = datos_4[0]*p
datos_4[-1] = datos_4[-1]*p

centrimedia = sum(datos_4)/(n-2*k)

print("La centrimedia es %s" %centrimedia)
```

b) Calcula la media winsorizada al 20 %.

```python
np.random.seed(seed=2026) #semilla para obtener el mismo resultado
datos= np.random.uniform(a,b,n)
datos.sort()

#Cálculo de la media winsorizada al 20%
alfa = 0.2
k = int(n*alfa)
datos_5 = datos

datos_5[0:k-1] = datos[k] #Se reemplazan los primeros k datos
datos_5[n-k:] = datos[n-1-k] #Se reemplazan los últimos k datos

media_winsorizada = np.sum(datos_5)/n

print("La media winzorizada al 20%% es %s" %media_winsorizada)
```

c) Ahora se econtrará la trimedia, para ello es necesario conocer el cuartil 1, la mediana y el cuartil 3.

```python
q1_x = np.quantile(datos, 0.25 ) #Cuartil 1
me_x = np.quantile(datos, 0.50 ) #Mediana
q3_x = np.quantile(datos, 0.75) #Cuartil 3

trimedia = (q1_x+2*me_x+q3_x)/4

print("La trimedia es %s" %trimedia)
```

d) Finalmente, se encontrará la media muestral y se realizará la comparativa

```python
media = np.mean(datos)

Medidas = ["Media muestral", "Media recortada 15%%", "Media recortada 20%%",
           "Centrimedia", "Media winzorizada 20%%", "Trimedia"]

valores_medidas = np.array([media, media_recortada_15, media_recortada_20,
                            centrimedia, media_winsorizada, trimedia])

resultados = pd.DataFrame()
resultados["Medidas"] = Medidas
resultados["Valores"] = valores_medidas
```
