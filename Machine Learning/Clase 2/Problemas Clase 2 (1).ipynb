```python
import random
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

# Ejercicio práctico

Se considera a $X$ como una variable que sigue una distribución $\text{Beta}(\alpha, \beta)$ cuya función de densidad es 
$$f(x;\alpha,\beta)=\dfrac{x^{\alpha-1}(1-x)^{\alpha}}{B(\alpha,\beta)}$$

Para determinar el valor de $\alpha$ y $\beta$ se considera el número de carnet, el penúltimo dígito será $\alpha$ y el último dígito será $\beta$. Por ejemplo si el carnet es $U20231085$, entonces $\alpha=8$ y $\beta=5$

a) Obtener una muestra de 10,000 valores.

```python
#Sugerencia usar
alfa = 8
beta = 5
n = 10000

np.random.seed(20231085)
x = np.random.beta(alfa,beta,n)
```

b) Graficar histograma y PDF teórica.

```python
#Sugerencia
y = np.linspace(0,1,100)
plt.figure()
plt.hist(x,bins=30,density=True,edgecolor="Black")
pdf = stats.beta.pdf(y, alfa, beta)
plt.plot(y, pdf)
plt.show()
```

c) Calcula media y varianza de la muestra

```python
media = np.mean(x)
var = np.var(x)

print("La media es %s" %media)
print("La varianza es %s" %var)
```

d) Comparar los resultados con la media y varianza exacta. La media y la varianza exacta se encuentran como

$$\mu = \dfrac{\alpha}{\alpha+\beta}, \quad \sigma^2 = \dfrac{\alpha \beta}{(\alpha+\beta+1)(\alpha+\beta)^2}$$

```python
media_exacta = alfa/(alfa+beta)
var_exacta = alfa*beta/((alfa+beta+1)*(alfa+beta)**2)

print("La media exacta es %s" %media_exacta)
print("La varianza exacta es %s" %var_exacta)
```
