## Ejercicio 1

El siguiente conjunto de datos sirve para analizar y predecir los precios de las acciones de una empresa. Para ambos problemas se pide:

<ul>
  <li>
    Describir el conjunto de datos aplicando los métodos <code>info()</code>, <code>describe()</code>, <code>head()</code> y <code>tail()</code>. Además, realizar un análisis gráfico de todas las variables, tanto predictoras como variable respuesta. Posteriormente, estandarizar las variables predictoras, lo cual es importante para aplicar los modelos de regresión Ridge, Lasso y Red Elástica.
  </li>

  <li>
    Dividir el conjunto de datos en dos partes: el 75% de las observaciones se utilizará para el entrenamiento del modelo y el 25% restante para la validación de los distintos métodos de regresión aplicados.
  </li>

  <li>
    Realizar una regresión de tipo Ridge utilizando los valores de regularización <code>alpha = 1</code> y <code>alpha = 1000</code>.
  </li>

  <li>
    Realizar una regresión de tipo Lasso utilizando los valores de regularización <code>alpha = 1</code> y <code>alpha = 10</code>.
  </li>

  <li>
    Aplicar validación cruzada para determinar el valor de <code>alpha</code> que proporciona el menor error cuadrático medio en los modelos de regresión Ridge y Lasso.
  </li>

  <li>
    Realizar una regresión mediante Red Elástica. Para ello, se deberán probar distintos valores de <code>r</code> y obtener los valores óptimos de <code>r</code> y <code>alpha</code> mediante validación cruzada.
  </li>

  <li>
    Comprobar, utilizando la partición de validación, cuál de los modelos obtiene el menor error cuadrático medio.
  </li>

  <li>
    Obtener una predicción utilizando la media de las variables predictoras con el modelo que haya presentado el menor error cuadrático medio.
  </li>
