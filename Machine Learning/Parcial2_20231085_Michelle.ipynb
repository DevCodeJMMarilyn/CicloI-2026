## 1. Generación de la semilla

### Explicación

En esta sección se trabajará con el código del estudiante para construir la semilla que será utilizada en la generación del conjunto de datos.

El procedimiento consiste en:
- tomar el código sin la letra U,
- revisar los últimos cuatro dígitos,
- y reemplazar cualquier valor menor que 2 por el número 2.


```
codigo = "20231085"
codigo
```

### Explicación

Se extraen los últimos cuatro dígitos del código del estudiante para verificar cuáles deben ser modificados según las instrucciones proporcionadas en la rúbrica.


```
ultimos = list(codigo[-4:])
ultimos
```

### Explicación

Los valores menores que 2 son reemplazados por el número 2 para construir correctamente la semilla utilizada en el problema de regresión.


```
ultimos_corregidos = []

for digito in ultimos:

    if int(digito) < 2:
        ultimos_corregidos.append("2")

    else:
        ultimos_corregidos.append(digito)

ultimos_corregidos
```

### Explicación

Después de realizar la corrección de los últimos cuatro dígitos, se construye la semilla final que será utilizada como random_state.


```
semilla = codigo[:-4] + "".join(ultimos_corregidos)

semilla
```

```
digitos = [int(x) for x in str(semilla)]

digitos
```

## 2. Construcción de parámetros

### Explicación

A partir de los dígitos de la semilla se calculan los parámetros solicitados en la rúbrica para generar el conjunto de datos.

Estos parámetros determinan:
- el número de observaciones,
- la cantidad de variables predictoras,
- el número de variables informativas,
- y el nivel de ruido del modelo.


```
n_samples = 900 + 10*(digitos[4])

n_features = 10 + digitos[6] + digitos[7]

n_informative = 10 + digitos[7]

bias = 2

noise = 10*(digitos[5])

random_state = int(semilla)

shuffle = False

print("n_samples =", n_samples)
print("n_features =", n_features)
print("n_informative =", n_informative)
print("bias =", bias)
print("noise =", noise)
print("random_state =", random_state)
print("shuffle =", shuffle)
```

## 3. Generación del conjunto de datos

### Explicación

Se utiliza la función make_regression() de la librería scikit-learn para generar un conjunto de datos sintético orientado a problemas de regresión múltiple.


```
import pandas as pd
import numpy as np
from sklearn.datasets import make_regression

X, y = make_regression(
    n_samples = n_samples,
    n_features = n_features,
    n_informative = n_informative,
    bias = bias,
    noise = noise,
    random_state = random_state,
    shuffle = shuffle
)
```

### Explicación

El conjunto de datos es convertido a un DataFrame para facilitar el análisis, manipulación y aplicación de los modelos de regresión.


```
columnas = []

for i in range(n_features):

    columnas.append("Variable_" + str(i+1))

datos = pd.DataFrame(X, columns = columnas)

datos["Respuesta"] = y

datos
```

## 8. Separación de variables

### Explicación

Se separa la variable respuesta de las variables predictoras para preparar el conjunto de datos antes del entrenamiento de los modelos.


```
respuesta = datos["Respuesta"].copy()

predictoras = datos.drop(["Respuesta"], axis = 1)

predictoras.head()
```

## 9. Estandarización de variables

### Explicación

La estandarización permite trabajar todas las variables en la misma escala.

Este procedimiento es importante en modelos penalizados como Ridge y Red Elástica.


```
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(predictoras)

predictoras_est = pd.DataFrame(
    scaler.transform(predictoras),
    columns = predictoras.columns
)

predictoras_est
```

## 10. División entrenamiento-validación

### Explicación

El conjunto de datos se divide en:
- 70% para entrenamiento,
- y 30% para validación,

tal como lo solicita la rúbrica.


```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    predictoras_est,
    respuesta,
    test_size = 0.30,
    random_state = int(semilla)
)
```

## 11. Regresión Ridge con alpha = 1

### Explicación

Se ajusta un modelo Ridge utilizando un nivel de penalización pequeño sobre los coeficientes.


```
from sklearn.linear_model import Ridge

ridge_reg = Ridge(alpha = 1,
                  solver = "auto")

ridge_reg.fit(X_train, y_train)

print("El intercepto que se tiene es %s" %ridge_reg.intercept_)

coefs_ridge_a = pd.DataFrame({
    "Variables": predictoras.columns,
    "Coeficientes": ridge_reg.coef_
})

coefs_ridge_a
```

## 12. Regresión Ridge con alpha = 50

### Explicación

Se incrementa el nivel de penalización para observar cómo cambian los coeficientes del modelo.


```
ridge_reg_b = Ridge(alpha = 50,
                    solver = "auto")

ridge_reg_b.fit(X_train, y_train)

print("El intercepto que se tiene es %s" %ridge_reg_b.intercept_)

coefs_ridge_b = pd.DataFrame({
    "Variables": predictoras.columns,
    "Coeficientes": ridge_reg_b.coef_
})

coefs_ridge_b
```

## 11. Regresión Ridge con alpha = 1

### Explicación

Se ajusta un modelo Ridge utilizando un nivel de penalización pequeño sobre los coeficientes.


```
ridge_reg_c = Ridge(alpha = 100,
                    solver = "auto")

ridge_reg_c.fit(X_train, y_train)

print("El intercepto que se tiene es %s" %ridge_reg_c.intercept_)

coefs_ridge_c = pd.DataFrame({
    "Variables": predictoras.columns,
    "Coeficientes": ridge_reg_c.coef_
})

coefs_ridge_c
```

## 14. Comparación de modelos Ridge

### Explicación

Se comparan los distintos modelos mediante el error cuadrático medio (MSE) para identificar cuál presenta mejor desempeño predictivo.


```
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lm = LinearRegression()

lm.fit(X_train, y_train)

y_pred_lm = lm.predict(X_test)

y_pred_ridge = ridge_reg.predict(X_test)

y_pred_ridge_b = ridge_reg_b.predict(X_test)

y_pred_ridge_c = ridge_reg_c.predict(X_test)

mse_lm = mean_squared_error(y_test, y_pred_lm)

mse_ridge = mean_squared_error(y_test, y_pred_ridge)

mse_ridge_b = mean_squared_error(y_test, y_pred_ridge_b)

mse_ridge_c = mean_squared_error(y_test, y_pred_ridge_c)

comparacion_ridge = pd.DataFrame({
    "Modelo": ["Lineal",
               "Ridge alpha=1",
               "Ridge alpha=50",
               "Ridge alpha=100"],

    "MSE": [mse_lm,
            mse_ridge,
            mse_ridge_b,
            mse_ridge_c]
})

comparacion_ridge
```

## 15. Gráfico de coeficientes Ridge

### Explicación

El gráfico permite analizar cómo cambian los coeficientes de los modelos conforme aumenta el valor de alpha.


```
fig = plt.figure(figsize=(12,6))

barWidth = 0.15

r1 = np.arange(len(lm.coef_))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]

plt.bar(r1, lm.coef_,
       width = barWidth,
       edgecolor = "white",
       label = "Lineal")

plt.bar(r2, ridge_reg.coef_,
       width = barWidth,
       edgecolor = "white",
       label = "Ridge alpha=1")

plt.bar(r3, ridge_reg_b.coef_,
       width = barWidth,
       edgecolor = "white",
       label = "Ridge alpha=50")

plt.bar(r4, ridge_reg_c.coef_,
       width = barWidth,
       edgecolor = "white",
       label = "Ridge alpha=100")

plt.xticks([r + barWidth for r in range(len(lm.coef_))],
          [predictoras.columns[i] for i in range(len(predictoras.columns))],
          rotation = 90)

plt.xlabel("Variables")
plt.ylabel("Coeficientes")

plt.legend()

plt.show(fig)
```

## 16. Validación cruzada para Ridge

### Explicación

La validación cruzada permite encontrar el valor de alpha que minimiza el error cuadrático medio del modelo Ridge.


```
from sklearn.linear_model import RidgeCV
from sklearn.model_selection import KFold, cross_val_score

alphas = np.logspace(-3, 4, 100)

ridge = RidgeCV(alphas = alphas,
                cv = 10)

ridge.fit(X_train, y_train)

kf = KFold(
    n_splits = 10,
    shuffle = True,
    random_state = int(semilla)
)

mse_mean = list()
mse_std = list()

for alpha in alphas:

    modelo_tmp = RidgeCV(alphas = [alpha], cv = None)

    scores = cross_val_score(
        modelo_tmp,
        X_train,
        y_train,
        scoring = 'neg_mean_squared_error',
        cv = kf
    )

    mse = -scores

    mse_mean.append(mse.mean())
    mse_std.append(mse.std())

mse_mean = np.array(mse_mean)

mse_std = np.array(mse_std)

alpha_min_ridge = ridge.alpha_
```

## 17. Gráfico de validación cruzada

### Explicación

El gráfico muestra el comportamiento del error cuadrático medio para distintos valores de alpha.


```
x = np.log(alphas)

plt.figure(figsize=(8,4))

plt.errorbar(
    x,
    mse_mean,
    yerr = mse_std,
    fmt = 'o',
    color = 'red',
    ecolor = 'gray'
)

plt.axvline(
    np.log(alpha_min_ridge),
    linestyle = '--',
    color = "gray"
)

plt.xlabel("log(alpha)")
plt.ylabel("Mean Squared Error")
plt.title("Validación Cruzada para Ridge")

plt.gca().invert_xaxis()

plt.tight_layout()

plt.show()
```

```
print("El valor de alpha que minimiza el MSE es %s" %alpha_min_ridge)
```

## 18. Regresión mediante Red Elástica

### Explicación

La Red Elástica combina características de Ridge y Lasso permitiendo aplicar distintos niveles de penalización.


```
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import ElasticNetCV

e_net_a = ElasticNet(
    alpha = 1,
    l1_ratio = 0.1,
    random_state = int(semilla)
)

e_net_a.fit(X_train, y_train)

y_pred_elastic_a = e_net_a.predict(X_test)

mse_elastic_a = mean_squared_error(
    y_test,
    y_pred_elastic_a
)
```

## 18. Regresión mediante Red Elástica

### Explicación

La Red Elástica combina características de Ridge y Lasso permitiendo aplicar distintos niveles de penalización.


```

e_net_b = ElasticNet(
    alpha = 1,
    l1_ratio = 0.5,
    random_state = int(semilla)
)

e_net_b.fit(X_train, y_train)

y_pred_elastic_b = e_net_b.predict(X_test)

mse_elastic_b = mean_squared_error(
    y_test,
    y_pred_elastic_b
)
```

## 18. Regresión mediante Red Elástica

### Explicación

La Red Elástica combina características de Ridge y Lasso permitiendo aplicar distintos niveles de penalización.


```
e_net_c = ElasticNet(
    alpha = 1,
    l1_ratio = 0.9,
    random_state = int(semilla)
)

e_net_c.fit(X_train, y_train)

y_pred_elastic_c = e_net_c.predict(X_test)

mse_elastic_c = mean_squared_error(
    y_test,
    y_pred_elastic_c
)
```

## 19. Validación cruzada para Red Elástica

### Explicación

Se aplicará validación cruzada para obtener los valores óptimos de r y alpha.


```
modelo_cv = ElasticNetCV(
    l1_ratio = [0.1,0.3,0.5,0.7,0.9,1],
    alphas = np.logspace(-3,4,100),
    cv = 10,
    random_state = int(semilla)
)

modelo_cv.fit(X_train, y_train)
```

## 18. Regresión mediante Red Elástica

### Explicación

La Red Elástica combina características de Ridge y Lasso permitiendo aplicar distintos niveles de penalización.


```
elastic_model = ElasticNet(
    alpha = modelo_cv.alpha_,
    l1_ratio = modelo_cv.l1_ratio_,
    random_state = int(semilla)
)

elastic_model.fit(X_train, y_train)

y_pred_elastic_cv = elastic_model.predict(X_test)

mse_elastic_cv = mean_squared_error(
    y_test,
    y_pred_elastic_cv
)

mse_elastic_cv
```

## 20. Comparación final de modelos

### Explicación

Se comparan todos los modelos utilizando el error cuadrático medio para identificar cuál presenta mejor capacidad predictiva.


```
comparacion_final = pd.DataFrame({
    "Modelo": ["Lineal",
               "Ridge alpha=1",
               "Ridge alpha=50",
               "Ridge alpha=100",
               "Red Elástica r=0.1",
               "Red Elástica r=0.5",
               "Red Elástica r=0.9",
               "Red Elástica CV"],

    "MSE": [mse_lm,
            mse_ridge,
            mse_ridge_b,
            mse_ridge_c,
            mse_elastic_a,
            mse_elastic_b,
            mse_elastic_c,
            mse_elastic_cv]
})

comparacion_final
```

## 21. Selección del mejor modelo

### Explicación

El mejor modelo será aquel que presente el menor error cuadrático medio sobre la partición de validación.


```
mejor_modelo = comparacion_final.iloc[
    comparacion_final["MSE"].idxmin(),0
]

menor_mse = comparacion_final["MSE"].min()

print("El mejor modelo es: %s" %mejor_modelo)

print("El menor MSE es: %s" %menor_mse)
```
