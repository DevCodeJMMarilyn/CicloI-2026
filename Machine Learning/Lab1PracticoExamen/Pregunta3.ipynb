```python
import numpy as np
from scipy import stats
from scipy.stats import mstats
np.random.seed(20231085)
```

### A) Datos de muestra A, B y C

```python
muestra_A = np.random.normal(loc=30, scale=5, size=180)
muestra_B = np.random.normal(loc=30, scale=15, size=20)

muestra_C = np.concatenate([muestra_A, muestra_B])

media_rec_15 = stats.trim_mean(muestra_C, 0.15)
centrimedia = stats.trim_mean(muestra_C, 0.25)

muestra_win = mstats.winsorize(muestra_C, limits=[0.10, 0.10])
media_win_10 = np.mean(muestra_win)
```

B) Determinar la media recortada al 15% y la centrimedia.

```python
media_rec_15 = stats.trim_mean(muestra_C, 0.15)
centrimedia = stats.trim_mean(muestra_C, 0.25)
```

C) Calcular la media winsorizada al 10%.

```python
muestra_win = mstats.winsorize(muestra_C, limits=[0.10, 0.10])
media_win_10 = np.mean(muestra_win)
```

D) Calcular la trimedia.

```python
q1, q2, q3 = np.percentile(muestra_C, [25, 50, 75])
trimedia = (q1 + 2 * q2 + q3) / 4
```

E) Determinar la media muestral y comparar con los resultados anteriores.

```python
media_muestral = np.mean(muestra_C) #media muestral
```

```python
print(f"Media Recortada 15%: {media_rec_15:.4f}")
print(f"Centrimedia: {centrimedia:.4f}")
print(f"Media Winsorizada 10%: {media_win_10:.4f}")
print(f"Trimedia: {trimedia:.4f}")
print(f"Media Muestral Simple: {media_muestral:.4f}")
```
