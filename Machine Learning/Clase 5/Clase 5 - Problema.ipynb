## Ejercicio propuesto

El conjunto de datos contiene el conteo de bicicletas públicas alquiladas por hora en el Sistema de Bicicletas Compartidas de Seúl, junto con los datos meteorológicos correspondientes y la información sobre días festivos.

```python
import numpy as np
import pandas as pd
```

```python
#Leer el archivo
datos =  pd.read_csv("SeoulBikeData.csv",encoding="latin1")
```

```python
#5 histogramas
import matplotlib.pyplot as plt

variables = ["Temperature(°C)", "Wind speed (m/s)", "Dew point temperature(°C)"] #Eligir variables

plt.figure(figsize=(10,6))

for i, var in enumerate(variables):
    plt.subplot(2, 2, i+1)
    plt.hist(datos[var], edgecolor="black",bins=15)
    plt.title(var)

plt.tight_layout()
plt.show()
```

```python
data_num = datos.drop(['Date', "Seasons", "Holiday", "Functioning Day"], axis=1)
#matriz
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

matrix = data_num.corr()
plt.figure(figsize=(10,8), dpi =500)
sns.heatmap(matrix,annot=True,fmt=".4f", linewidth=.5)
plt.show()
```
