````python
#Colocar variables numéricas
import matplotlib.pyplot as plt
variables = ["age", "social_media_hours", "work_or_study_hours",  "entertainment_hours",  "total_screen_time"]

plt.figure(figsize=(12,8))

for i, var in enumerate(variables):
    plt.subplot(2, 3, i+1)
    plt.hist(data_set[var], edgecolor="black",bins=10)
    plt.title(var)

plt.tight_layout()
plt.show()
````

La matriz de convariaza indica que tan correlacionadas están dos variables numéricas. Cada entrada de la matriz es un valor entre $-1$ y $1$. Obtener la matriz de convariaza para las variables cuantitativas.

Ahora, es necesario eliminar las variables categóricas y quedarnos solamente con las variables cuantitativas.

````python
data_num = data_set.drop(['date',"user_id", "age_group", "primary_device","internet_type"], axis=1) 
#Eliminar variables categóricas
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

matrix = data_num.corr()
plt.figure(figsize=(10,8), dpi =500)
sns.heatmap(matrix,annot=True,fmt=".4f", linewidth=.5)
plt.show()
````

Realizar el diagrama de dispersión para las variables que presentan un mayor coeficiente de correlación en la matriz de covarianza.

````python
#Las variables que mayor correlación tienen son total_screen_time y social_media_hours
sns.scatterplot(data=data_set, x="work_or_study_hours", y="total_screen_time")
````
