Al estudiar estos datos, se pueden identificar patrones, comparar resultados y obtener conclusiones que ayuden a mejorar hábitos de sueño saludables.

Realizar lo siguiente:

f) Realizar histogramas para las variables body_weight,	brain_weight,max_life_span y total_sleep.

## Solución

a) Se usará el comando pd.read_csv para cargar el archivo y luego mostrar el conjunto de datos.

```python
import pandas as pd

datos = pd.read_csv("Horas_de_sueño-clase 5.csv")
datos
```

b) Se mostrarán las primeras y últimas 10 observaciones, para ellos se usan los métodos .head() y .tail()

```python
datos.head(10)
```

```python
datos.tail(10)
```

c) Para obtener una descripción de las variables que conforman la base de datos se usará el método .info()

```python
datos.info()
```

Se puede observar que hay 8 variables en total, pero max_life_span, gestation_time y total_sleep aparecen como tipo objeto a pesar de que el DataFrame se registran valores numéricos

```python
#Mostrar los estadísticos
datos.describe()
```

d) Para que las variables ya no sean identificadas como object a pesar de ser numéricas, se reemplazarán las que aparecen con ? y en su lugar se indicará que es un valor nulo. Por lo que trabajamos con las columnas que tienen ese incoveniente.

```python
#datos["body_weight"] = pd.to_numeric(datos["body_weight"], errors = "coerce")
datos["max_life_span"] = pd.to_numeric(datos["max_life_span"], errors="coerce")
datos["gestation_time"] = pd.to_numeric(datos["gestation_time"], errors="coerce")
datos["total_sleep"] = pd.to_numeric(datos["total_sleep"], errors="coerce")
```

```python
datos.info()
```

Podemos observar que ya se observa el análisis para todas las variables numéricas, sin embargo en las columnas max_life_span, gestation_time y total_sleep hay únicamente 58 registros debido a la utilización de valores nulos.

f) Ahora se realizarán los histogramas de las variables indicadas, para ello se extraen en una lista para graficarlas de forma más eficaz

```python
import matplotlib.pyplot as plt

variables = ["body_weight", "brain_weight", "gestation_time","total_sleep"] #Eligir variables

plt.figure(figsize=(10,6))

for i, var in enumerate(variables):
    plt.subplot(2, 2, i+1)
    plt.hist(datos[var], edgecolor="black",bins=15)
    plt.title(var)

plt.tight_layout()
plt.show()
```

g) Ahora se realizará el gráfico de dispersión, para ello se usará sns.scatterplot junto con las variables de interés a graficar.

```python
import seaborn as sns

#Las variables son gestation_time y total_sleep
plt.figure()
sns.scatterplot(data=datos, x="gestation_time",y="total_sleep")
plt.show()
```

Se puede observar que a medida que el tiempo de gestación aumenta hay una disminución en el tiempo total de sueño. Lo que indicar que las variables están relacionadas de forma inversa.
