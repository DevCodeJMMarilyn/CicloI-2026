## 1. Regresión Lineal Simple

```
# PASO 1: Graficar datos
plt.plot(x, y, '*')
```

```
# PASO 2: Crear modelo
from sklearn.linear_model import LinearRegression
modelo = LinearRegression()
modelo.fit(x.reshape(-1,1), y)
```

```
# PASO 3: Obtener parámetros
b1 = modelo.coef_
b0 = modelo.intercept_

print("Ecuación:")
print(f"y = {b0} + {b1}x")
```

```
# PASO 4: Predicciones
y_pred = modelo.predict(x.reshape(-1,1))
modelo.predict([[5]])
```

```
# PASO 5: Métricas
from sklearn.metrics import mean_squared_error
import numpy as np

mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
r2 = modelo.score(x.reshape(-1,1), y)

print(mse, rmse, r2)
```

```
# PASO 6: Gráfica con recta
plt.scatter(x,y)
plt.plot(x, y_pred)
```

```
# PASO 7: Residuos
residuos = y - y_pred
plt.scatter(x, residuos)
```

```
# PASO 8: Normalidad
from scipy import stats
stats.shapiro(residuos)
```

```
# PASO 9: Homocedasticidad
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan

exog = sm.add_constant(x)
het_breuschpagan(residuos, exog)
```

```
# PASO 10: Intervalo confianza
n = len(x)
s2 = sum((y - y_pred)**2)/(n-2)
sb1 = np.sqrt(s2/(np.var(x)*n))

IC_inf = b1 - 1.96*sb1
IC_sup = b1 + 1.96*sb1

print(IC_inf, IC_sup)
```
