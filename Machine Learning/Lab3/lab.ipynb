Transformar variables categóricas

```python
print(df.columns.tolist())
```

```python
df = pd.get_dummies(df,
                    columns=['Gender',
                             'Marital_Status',
                             'Department',
                             'Job_Role',
                             'Overtime'],
                    drop_first=True)
```

```python
sns.countplot(
    x='Overtime_Yes',
    hue='Attrition',
    data=df
)

plt.title('Horas extra y abandono')
plt.show()
```

```python
#definimos x y Y
X = df.drop(['Employee_ID',
             'Attrition'], axis=1)

y = df['Attrition']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.30,
    random_state=2026
)
```

```python
plt.figure(figsize=(15,10))

sns.heatmap(
    df.corr(numeric_only=True),
    cmap='coolwarm'
)

plt.title('Matriz de correlación')
plt.show()
```

```python
#Estandarizacion
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)
```

```python
#Regresion logistica
from sklearn.linear_model import LogisticRegression

modelo_log = LogisticRegression(max_iter=1000)

modelo_log.fit(X_train_scaled,
               y_train)

pred_log = modelo_log.predict(X_test_scaled)
```

```python
#Evaluación Regresion Logistica
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

print("Accuracy")

print(accuracy_score(y_test,
                     pred_log))

print(confusion_matrix(y_test,
                       pred_log))

print(classification_report(y_test,
                            pred_log))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test,pred_log)

sns.heatmap(cm,
            annot=True,
            fmt='d')

plt.title('Matriz de Confusión - Regresión Logística')
plt.ylabel('Real')
plt.xlabel('Predicción')

plt.show()
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train_scaled,
        y_train)

pred_knn = knn.predict(X_test_scaled)
```

```python
#Evaluamos knn
print("Accuracy KNN")

print(accuracy_score(y_test,
                     pred_knn))

print(confusion_matrix(y_test,
                       pred_knn))

print(classification_report(y_test,
                            pred_knn))
```

```python
cm = confusion_matrix(y_test,pred_knn)

sns.heatmap(cm,
            annot=True,
            fmt='d')

plt.title('Matriz de Confusión - KNN')
plt.ylabel('Real')
plt.xlabel('Predicción')

plt.show()
```

```python
#Arbol de decision
from sklearn.tree import DecisionTreeClassifier

arbol = DecisionTreeClassifier(
    random_state=123
)

arbol.fit(X_train,
          y_train)

pred_arbol = arbol.predict(X_test)
```

```python
cm = confusion_matrix(y_test,pred_arbol)

sns.heatmap(cm,
            annot=True,
            fmt='d')

plt.title('Matriz de Confusión - Árbol de Decisión')
plt.ylabel('Real')
plt.xlabel('Predicción')

plt.show()
```

```python
print("Accuracy Árbol")

print(accuracy_score(y_test,
                     pred_arbol))

print(confusion_matrix(y_test,
                       pred_arbol))

print(classification_report(y_test,
                            pred_arbol))
```

```python
#Comparacion
print("Regresión Logística:",
      accuracy_score(y_test,pred_log))

print("KNN:",
      accuracy_score(y_test,pred_knn))

print("Árbol:",
      accuracy_score(y_test,pred_arbol))
```

```python
accuracy_log = accuracy_score(y_test,pred_log)
accuracy_knn = accuracy_score(y_test,pred_knn)
accuracy_arbol = accuracy_score(y_test,pred_arbol)

modelos = ['Logística',
           'KNN',
           'Árbol']

accuracy = [accuracy_log,
            accuracy_knn,
            accuracy_arbol]

plt.figure(figsize=(8,5))

plt.bar(modelos,accuracy)

plt.title('Comparación de Accuracy')
plt.ylabel('Accuracy')

plt.show()
```
