### F. Representar mediante histogramas las variables cuantitativas.

```python
variables = [
    "job_id",
    "year",
    "automation_risk_percent",
    "salary_after_usd",
    "salary_change_percent",
    "skill_demand_growth_percent",
    "remote_feasibility_score",
    "ai_replacement_score",
    "skill_gap_index",
    "salary_before_usd",
    "ai_adoption_level",
    "education_requirement_level",
    "skill_transition_pressure",
    "wage_volatility_index",
    "reskilling_urgency_score",
    "ai_disruption_intensity"
]
plt.figure(figsize=(6,6))
for v in variables:
    plt.hist(df[v], bins=20 , edgecolor= "black", color= "purple")
    plt.title("Histograma de " + v)
    plt.xlabel(v)
    plt.ylabel("Frecuencia")
    plt.show()
```

### G. Realizar diagramas de caja y bigote para indicar las variables que presentan datos atípicos.

```python
vars_cuantitativas = df.select_dtypes(include=[np.number]).columns
# Diagramas de caja y bigote
fig, axes = plt.subplots(nrows=int(np.ceil(len(vars_cuantitativas)/4)), ncols=4, figsize=(18, 15))
axes = axes.flatten()
for i, col in enumerate(vars_cuantitativas):
    sns.boxplot(y=df[col], ax=axes[i], color='magenta')
    axes[i].set_title(col)
    axes[i].set_ylabel('')
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])
plt.suptitle("Diagramas de Caja y Bigote (Atípicos)", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

### H. Calcular la media recortada al 10 % para las variables crecimiento porcentual en la demanda de habilidades y porcentaje de cambio en el salario. Comparar los resultados con la media muestral.

```python
vars_recorte = ['skill_demand_growth_percent', 'salary_change_percent']
for var in vars_recorte:
    media_muestral = df[var].mean()
    media_recortada = stats.trim_mean(df[var].dropna(), 0.10)
    print(f"Variable: {var}")
    print(f"   Media Muestral:      {media_muestral:.4f}")
    print(f"   Media Recortada 10%: {media_recortada:.4f}")
```

### J.  Realizar la matriz de correlación para analizar la correlación lineal entre variables. Determinar las que tienen mayor correlación y algunas que no están correlacionadas.

```python
matriz_corr = df[vars_cuantitativas].corr()

plt.figure(figsize=(14, 10))
sns.heatmap(matriz_corr, annot=True, fmt=".2f", cmap='magma', vmin=-1, vmax=1)
plt.title("Matriz de Correlación Lineal")
plt.show()
```

### K.Crear un gráfico de dispersión para los tres pares de variables que tengan una mayor correlación en la matriz anterior.

```python
matriz_corr_abs = matriz_corr.abs()
upper_tri = matriz_corr_abs.where(np.triu(np.ones(matriz_corr_abs.shape), k=1).astype(bool))
pares_ordenados = upper_tri.unstack().dropna().sort_values(ascending=False)

top_3_pares = pares_ordenados.head(3).index.tolist()
peores_pares = pares_ordenados.tail(3).index.tolist()

print("Pares con mayor correlación lineal:")
for par in top_3_pares:
    print(f"{par[0]} y {par[1]} (r = {matriz_corr.loc[par[0], par[1]]:.4f})")
print("")
print("Pares no correlacionados")
for par in peores_pares:
    print(f"{par[0]} y {par[1]} (r = {matriz_corr.loc[par[0], par[1]]:.4f})")
```

```python
# Gráficos de dispersión
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
for i, par in enumerate(top_3_pares):
    sns.scatterplot(data=df, x=par[0], y=par[1], ax=axes[i], alpha=0.5, color='darkviolet')
    axes[i].set_title(f"{par[0]} vs {par[1]}")
plt.tight_layout()
plt.show()
```

### L. Con base a los resultados anteriores determinar los factores que inciden directamente en el impacto de la IA en el empleo.  Y cuales no tienen influencia significativa. Justificar su respuesta.

```python
print("Factores que inciden directamente en el impacto de la IA en el empleo")
corr_impacto = matriz_corr['ai_disruption_intensity'].drop('ai_disruption_intensity')

print("1. Factores con influencia directa significativa:")
factores_directos = corr_impacto[corr_impacto.abs() >= 0.3]
for var, val in factores_directos.sort_values(ascending=False).items():
    print(f"   - {var} (r = {val:.4f})")
```

**Respuesta: Estas variables presentan un coeficiente de correlación moderado a fuerte alejado de 0 y esto demuestra que a medida que estas métricas cambian el riesgo de automatización, la intensidad provocada por la IA se incrementa o disminuye de forma proporcional y predecible.**

```python
print("\n2. Factores sin influencia significativa:")
factores_nulos = corr_impacto[corr_impacto.abs() < 0.1]
for var, val in factores_nulos.sort_values(ascending=False).items():
    print(f"   - {var} (r = {val:.4f})")
```

**Respuesta: Estas variables presentan un coeficiente de correlación muy cercano a 0, lo que indica en estadistica que no existe una relación lineal con la intensidad de la disrupción por IA. Variaciones en estos aspectos no alteran de forma evidente el impacto global del reemplazo laboral**
