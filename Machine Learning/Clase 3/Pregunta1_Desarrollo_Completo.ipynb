## Enunciado

X₁ ~ Beta(2,3), n=2000  
X₂ ~ Normal(4,2), n=6000  
X₃ ~ Exponencial(λ=5), n=2000  
X₄ ~ Normal(3,1), n=6000  

Construir:

X = X₁ + X₃ + X₁X₃  
Y = 2X₂ − 3X₄ + 4  
Z = (Y − 3)/5  

Comprobar normalidad de X y Z.  
Realizar gráfico Q-Q de X y Y.



## 1️⃣ Fijar semilla (cambiar por su carnet)


```

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.api as sm

np.random.seed(123456)  # CAMBIAR POR SU CARNET
```

## 2️⃣ Generación de muestras

```

X1 = np.random.beta(2, 3, 2000)
X3 = np.random.exponential(1/5, 2000)
X2 = np.random.normal(4, 2, 6000)
X4 = np.random.normal(3, 1, 6000)
```

## 3️⃣ Construcción de nuevas variables

```

X = X1 + X3 + X1*X3
Y = 2*X2 - 3*X4 + 4
Z = (Y - 3)/5
```

## 4️⃣ Prueba de normalidad (Shapiro-Wilk)

```

print("p-valor X:", stats.shapiro(X).pvalue)
print("p-valor Z:", stats.shapiro(Z).pvalue)
```

## 5️⃣ Gráficos Q-Q

```

sm.qqplot(X, line='45')
plt.title("Q-Q plot X")
plt.show()

sm.qqplot(Y, line='45')
plt.title("Q-Q plot Y")
plt.show()
```


## 6️⃣ Conclusiones esperadas

X no debería ser normal (mezcla no lineal).  
Y sí es normal (combinación lineal de normales).  
Z también es normal (transformación lineal de Y).
