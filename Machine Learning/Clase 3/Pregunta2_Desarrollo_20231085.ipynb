## Datos del problema

K₁ ~ Beta(2,4), n=1000  
K₂ ~ Uniforme[-6,6], n=1000  

Y₁ = 200 + 150K₁ + 50K₂  
Y₂ = 200 + 150K₁ + 10K₂(2 + √K₁)

Se ajusta modelo: Y = β₀ + β₁K₁


```

import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan

np.random.seed(20231085)
```

```

K1 = np.random.beta(2, 4, 1000)
K2 = np.random.uniform(-6, 6, 1000)
```

```

Y1 = 200 + 150*K1 + 50*K2
Y2 = 200 + 150*K1 + 10*K2*(2 + np.sqrt(K1))
```

## Regresión Y1

```

X = sm.add_constant(K1)
modelo1 = sm.OLS(Y1, X).fit()
res1 = modelo1.resid
print(modelo1.summary())
```

```

plt.scatter(K1, res1)
plt.title("Residuos Y1")
plt.show()

print("Breusch-Pagan Y1 p-value:", het_breuschpagan(res1, X)[1])
```

## Regresión Y2

```

modelo2 = sm.OLS(Y2, X).fit()
res2 = modelo2.resid
print(modelo2.summary())
```

```

plt.scatter(K1, res2)
plt.title("Residuos Y2")
plt.show()

print("Breusch-Pagan Y2 p-value:", het_breuschpagan(res2, X)[1])
```


## Interpretación

Y₁ → homocedasticidad esperada.  
Y₂ → heterocedasticidad esperada.
